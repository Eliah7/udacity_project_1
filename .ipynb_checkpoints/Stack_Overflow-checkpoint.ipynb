## Qn 1: What were the most widely used programming languages among developers in the Stack Overflow Developer Survey, 2017?

In this analysis, we aimed to uncover the programming language landscape within the Stack Overflow Developer Survey, 2017. Identifying the most widely used languages provides valuable insights into the preferences and trends shaping the developer community during that particular year.

```python
df = data.copy()
df_qn1 = df[['Respondent', 'Professional', 'HaveWorkedLanguage', 'FormalEducation', 'YearsProgram', 'Country']]
```

## Data Understanding

```python
df_qn1.head()
```

### Handle Missing Data

```python
df_qn1.shape
```

```python
df_qn1.isnull().sum()
```

```python
## Drop all respondents who have no working language
df_qn1 = df_qn1.dropna()
```

```python
df_qn1.shape
```

### Data Processing 

```python
## Define if a respondent is a professional
df_qn1.Professional.unique()
```

```python
df_qn1['IsProfessional'] = df_qn1['Professional'].apply(lambda x: 1 if 'professional' in x.lower() else 0 )
```

```python
df_qn1.head()
```

## Prepare Data

```python
languages = []
```

```python
for lang_list in df_qn1.HaveWorkedLanguage.apply(lambda x: x.strip().split(";")).to_list():
    languages +=  lang_list
#     languages.append(lang_list)
```

```python
languages = set([lang.replace(" ", '') for lang in languages])
```

```python
languages
```

```python
for lang in languages:
    # create a column in dictionary
    df_qn1[lang] = 0
    df_qn1[lang] = df_qn1.HaveWorkedLanguage.apply(lambda x: 1 if lang.lower() in x.lower() else 0 )
```

```python
round(df_qn1.describe(), 2)
```

```python
df_qn1.head()
```

### Evaluate the Results

```python
plt.figure(figsize=(10, 5))
plt.title('Most used Languages')
ax = df_qn1[languages].sum().plot(kind='bar')
```

## Qn 2. Are there any correlations between factors such as age, education, or years of experience with the technologies developers choose to use?

Exploring the intricate relationships between demographic factors and technology preferences is key to understanding the nuanced decisions made by developers. This question delves into correlations, unveiling how aspects like age, education, and years of experience influence the selection of programming languages and technologies.

```python
df_qn2 = df_qn1.copy()
```

### Data Understanding

```python
df_qn2.head()
```

### Handle Missing Data

```python
df_qn2.shape
```

```python
df_qn2.isnull().sum()
```

```python
# There are no missing data
```

### Data Processing 

```python
# Handling Categorical Columns
# FormalEducation 
df_qn2.FormalEducation.unique()
```

```python
## drop all 'I prefer not to answer'
mask = df_qn2.FormalEducation == 'I prefer not to answer'

# select all rows except the ones that contain 'Coca Cola'
df_qn2 = df_qn2[~mask]
df_qn2.shape
```

```python
def ProcessFormalEducationCats(level):
    """
    Function to process Formal Education categorical values.
    """
    if level == 'I never completed any formal education':
        return 0
    elif level == 'Primary/elementary school':
        return 1
    elif level == 'Secondary school':
        return 2
    elif level == 'Some college/university study without earning a bachelor\'s degree':
        return 3
    elif level == 'Professional degree':
        return 4
    elif level == 'Bachelor\'s degree':
        return 5
    elif level == 'Master\'s degree':
        return 6
    elif level == 'Doctoral degree':
        return 7
```

```python
df_qn2['FormalEducationCat'] = df_qn2.FormalEducation.apply(ProcessFormalEducationCats)
```

```python
df_qn2.head()
```

```python
# Handling Categorical Columns
# YearsProgram
df_qn2.YearsProgram.unique()
```

```python
YearsProgramMask = {
    '2 to 3 years' : 2, 
    '9 to 10 years' : 9, 
    '20 or more years' : 20,
    '14 to 15 years' : 14, 
    '6 to 7 years' : 6, 
    '10 to 11 years' : 10,
    '13 to 14 years' : 13, 
    '12 to 13 years' : 12, 
    '15 to 16 years' : 15,
    '11 to 12 years' : 11, 
    '8 to 9 years' : 8, 
    '7 to 8 years' : 7, 
    '5 to 6 years' : 5,
    'Less than a year' : 0, 
    '4 to 5 years' : 4, 
    '3 to 4 years' : 3,
    '17 to 18 years' : 17, 
    '1 to 2 years' :  1, 
    '16 to 17 years' : 16,
    '18 to 19 years' : 18, 
    '19 to 20 years' : 19
}
```

```python
df_qn2['YearsProgramCat'] = df_qn2.YearsProgram.apply(lambda x: YearsProgramMask[x])
```

```python
df_qn2.head()
```

### Evaluate the Results

```python
plt.figure(figsize=(30, 30))
plt.title('Correlation between Education, Years of Experience and Technoloies developers choose to use')
sns.heatmap(df_qn2.corr(), cmap="YlGnBu", annot=True) 
```
